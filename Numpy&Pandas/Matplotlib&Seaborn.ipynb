```python
plt.scatter(x,y,c='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2D')
```

```python
x = np.arange(0,10)
y = np.arange(990,1000)
plt.plot(x,y)
```

```python
y=x*x
plt.plot(x,y,c='c')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('This is how you write title')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r-o')
plt.subplot(2,2,2)
plt.plot(x,y,'c-')
plt.subplot(2,2,4)
plt.plot(x,y,'g*')
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)

plt.title("Sine wave form")
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.title("Sine")
plt.plot(x,y_sin,'r')

plt.subplot(2,1,2)
plt.title("Cosine")
plt.plot(x,y_cos,'g')
plt.show()
```

## **Seaborn**

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.corr() # there is correlation between numeric columns only.
```

```python
sns.heatmap(df.corr()) # - helps in feature elimination
```

## **JoinPlot**
<br> A join plot allows to study the relationship between 2 numeric variables.

**Univariate & Bivariate Analysis**

```python
sns.jointplot(x='tip', y='total_bill', data = df, kind = 'hex') # try 'reg' for kind
```

## **Pairplot**

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='day')
```

## **Distplot**
<br> dist plot helps check the distribution of a column/feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

## Categorical Plots <br>
helps in analysing categorical data points. <br>
boxplot <br>
violinplot <br>
countplot <br>
barplot 

**CountPlot** <br> shows the counts of observations in each categorical bin of a feature using bars

```python
sns.countplot('sex', data=df) # give either x or y value and get its count
```

```python
sns.countplot('smoker', data=df, hue='sex')
```

**BarPlot**

```python
sns.barplot(x='total_bill',y='sex',data=df) # specify both x and y value 
```

**Boxplot** <br> box plot is a plot that represents min, 25%, 50%, 75%, max values of a variable/feature.

```python
sns.boxplot('day','total_bill', data=df) # the box plot also shows the outliers
```

```python
sns.boxplot( data=df) # not required to specify the x and y labels
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker', data=df)
```

```python
sns.violinplot(x='total_bill', y='day',data=df )
```
