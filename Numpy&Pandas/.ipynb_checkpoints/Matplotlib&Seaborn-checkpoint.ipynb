## Matplotlib

```python
plt.scatter(x,y,c='b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2D')
```

```python
x = np.arange(0,10)
y = np.arange(990,1000)
plt.plot(x,y)
```

```python
y=x*x
plt.plot(x,y,c='c')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('This is how you write title')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'ro')
plt.subplot(2,2,2)
plt.plot(x,y,'c-')
plt.subplot(2,2,4)
plt.plot(x,y,'g*')
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)

plt.title("Sine wave form")
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.title("Sine")
plt.plot(x,y_sin,'r')

plt.subplot(2,1,2)
plt.title("Cosine")
plt.plot(x,y_cos,'g')
plt.show()
```

```python
x=[2,8,10]
y=[11,16,9]



x2=[3,9,25]
y2=[6,16,64]
plt.bar(x,y)
plt.bar(x2,y2, color='g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Bar Graph')
plt.show()
```
